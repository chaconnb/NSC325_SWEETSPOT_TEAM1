```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
%matplotlib inline
```

```python
df = pd.read_csv("sweetspot_UT_Austin.csv")
print(df.shape)
print(df.columns)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Well Name','Bottomhole X', 'Bottomhole Y','TD(MD)'], axis = 1, inplace=True)
df.head()
#dropping the features we belieave that are associate to the process of indentifying the best places to drill well
```

```python
#check for missing values
df.isnull().sum()
#no missing values in our Sweet Spot data
```

```python
print(df.describe())
```

```python
zero_rows_df = df[df.eq(0).any(axis=1)]
print(zero_rows_df)
```

```python
print(df['Cw (bbl)'].describe())
```

```python
df['Cw (bbl)'].skew()
# highly positively skewed distribution, suggesting potential outliers or a log-normal type distribution.
```

```python
plt.figure(figsize= (10,8))
sns.distplot(df["Cw (bbl)"])
plt.title("Distribution of Cw (bbl)")
plt.show()
```

```python
df['Cw (bbl)'].kurt()
```

```python
correlation = df.corr()
correlation['Cw (bbl)'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(16,12))
plt.title('Correlation of Attributes with Cw (bbl)')
a = sns.heatmap(correlation, square=True, annot=True, fmt='.2f', linecolor='white')
a.set_xticklabels(a.get_xticklabels(), rotation=90)
a.set_yticklabels(a.get_yticklabels(), rotation=30)           
plt.show()
```
